```python
class Neuron:
    def __init__(self, xtrain, ytrain, learning_rate, epochs):
        self.xtrain = xtrain
        self.ytrain = ytrain
        self.learning_rate = learning_rate
        self.len_train = len(xtrain)
        self.epochs = epochs
        print(xtrain)
        print(self.len_train)
        
    def parameters_initializer(self):
        self.weight = 1
        self.bias = 1
        print("initial weight and bias:",self.weight,self.bias)
        
    def activate(self):
        self.parameters_initializer()
        epoch = 0
        while epoch<self.epochs and self.calculate_cost_function_mse()!=0:
            epoch+=1
            print("Epoch:",self.epochs,end="\n")
            print("revalued weight and bias",self.revalue_parameters(), end="\n----------------------------------\n")
    
    def calculate_cost_function_mse(self):
        cost_function_mse = 0
        for i in range(self.len_train):
            error=(self.ytrain[i][0]-self.calculate_ypred(self.xtrain[i][0]))
            cost_function_mse+=(error**2)
        print("cfe:",cost_function_mse)
        return cost_function_mse
    
    def calculate_partial_derivative_of_cost_function_withrespect_to_weight(self, old_bias):
        pd_cf_wrt_weight = 0
        for i in range(self.len_train):
            pd_cf_wrt_weight+=((-2*self.xtrain[i][0])*(self.ytrain[i][0]-(self.weight*self.xtrain[i][0]+old_bias)))
        print("pd_cf_wrt_weight: ",pd_cf_wrt_weight)
        return pd_cf_wrt_weight
            
    def calculate_partial_derivative_of_cost_function_withrespect_to_bias(self):
        pd_cf_wrt_bias = 0
        for i in range(self.len_train):
            pd_cf_wrt_bias+=(-2*(self.ytrain[i][0]-(self.weight*self.xtrain[i][0]+self.bias)))
        print("pd_cf_wrt_bias: ",pd_cf_wrt_bias)
        return pd_cf_wrt_bias
    
    def revalue_parameters(self):
        old_bias = self.bias
        self.bias = self.bias - (self.learning_rate*self.calculate_partial_derivative_of_cost_function_withrespect_to_bias())
        self.weight = self.weight - (self.learning_rate*self.calculate_partial_derivative_of_cost_function_withrespect_to_weight(old_bias))
        return self.weight, self.bias
    
    def calculate_ypred(self, x):
        y_pred = (self.weight*x+self.bias)
        print("y_pred:", y_pred)
        return y_pred


neuron = Neuron(x,y,0.01,50)
neuron.activate()
```
