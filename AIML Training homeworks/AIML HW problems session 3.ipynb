```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import itertools
from sklearn.neighbors import KNeighborsClassifier
```

```python
plt.figure(figsize=(5,5))
sns.scatterplot(data=df,x='Feature1',y='Feature2',hue="Class")
plt.title('Relationship between Feature1 and Feature2',size=15)
```

#### 4. Consider a collection of text documents (say 20 text files of ~1K size).  Compute the distribution of vowels, newlines, spaces and punctuations.  For example, if the text is "i am learning ml & ai", there are 21 letters and of those, there are 3 a, 1 e, 2 i, 0 o, 0u, 5 spaces, 0 newlines, and 1 punct.  The distribution is 3/21, 1/21, 2/21, 0/21, 0/21, 5/21, 0/21, 1/21, 9/21.


```python
import os
import glob
import sys

folder_path = "C:/Users/keerthivasan.s/Documents/TextFiles"
folder_path += "" if folder_path[-1]=="/" else "/"

txt_files = glob.glob(folder_path+"*.txt")
a,e,i,o,u,spaces,newlines,punct = 0,0,0,0,0,0,0,0
total=0
for file_path in txt_files:
    file = open(file_path, 'r', encoding='utf-8')
    for letter in file.read():
        letter = letter.lower()
        if letter == 'a':
            a+=1
        elif letter == 'e':
            e+=1
        elif letter == 'i':
            i+=1
        elif letter == 'o':
            o+=1
        elif letter == 'u':
            u+=1
        elif letter == ' ':
            spaces+=1
        elif letter == '\n':
            newlines+=1
        elif letter.isalpha()==0:
            punct+=1
        total+=1

print("Probablity Distribution of letter a: ", end="")
print("/".join(map(str,[a,total])))
print("Probablity Distribution of letter e: ", end="")
print("/".join(map(str,[e,total])))
print("Probablity Distribution of letter i: ", end="")
print("/".join(map(str,[i,total])))
print("Probablity Distribution of letter o: ", end="")
print("/".join(map(str,[o,total])))
print("Probablity Distribution of letter u: ", end="")
print("/".join(map(str,[u,total])))
print("Probablity Distribution of spaces: ", end="")
print("/".join(map(str,[spaces,total])))
print("Probablity Distribution of newlines: ", end="")
print("/".join(map(str,[newlines,total])))
print("Probablity Distribution of punctuations: ", end="")
print("/".join(map(str,[punct,total])))
```
